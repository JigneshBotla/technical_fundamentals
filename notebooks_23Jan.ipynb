```python
df = pd.read_csv('sales_large.csv',parse_dates=['order_date'])
df
```

```python
delivered = df[df['status']=='Delivered']
delivered
```

```python
filtered = df[(df['category']=='Electronics') & (df['region']=='North')&(df['status']=='Delivered')]
filtered
```

```python
df['revenue'] = df['price']*df['quantity']
df
```

```python
df['shipping_fee'] = df['revenue'].apply(lambda x:500 if x<5000 else 0)
df
```

```python
summary = df.groupby('category').agg({'revenue':['sum','mean'],'quantity':'sum'})
summary
```

```python
summary = summary.reset_index()
summary
```

```python
top_products = df.groupby('product')['revenue'].sum().sort_values(ascending=False)
top_products
```

```python
df['month']= df['order_date'].dt.to_period('M')
monthly_sales = df.groupby('month')['revenue'].sum().sort_values(ascending=False)
monthly_sales
```
