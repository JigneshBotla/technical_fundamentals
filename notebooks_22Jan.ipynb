```python
class BankAccount:
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance

    def deposit(self,amount):
        self.balance+=amount
        print("Amount deposited :",amount)

    def withdraw(self,amount):
        if amount<=self.balance:
            self.balance-=amount
            print("withdrawn amount :",amount)
        else:
            print("Insufficient balance")

    def show_balance(self):
        print("Account Holder :",self.name)
        print("Balance :",self.balance)

acc1 = BankAccount("Jignesh",50000)

acc1.deposit(1)

acc1.withdraw(2)

acc1.show_balance()

acc1.withdraw(50000)
    
```

```python
#E-Commerce order system

class Order:
    def __init__(self,*args):
        self.customer_name = args[0]
        self.products = []
        self.total_amount=0

    def add_products(self,product_name,price):
        self.products.append(product_name)
        self.total_amount+=price

    def apply_discount(self,percentage):
        discount = self.total_amount*(percentage/100)
        self.total_amount-=discount

    def generate_invoice(self):
        print("Customer :",self.customer_name)
        print("Products :",self.products)
        print("Total Amount :",self.total_amount)

order1 = Order("Jignesh")
order1.add_products("Towel",500)
order1.add_products("Bedsheet",1500)
order1.apply_discount(10)
order1.generate_invoice()
```

```python
def my_decorator(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")

    return wrapper


@my_decorator
def greet():
    print("Hello User")

greet()
```

```python
# Decorator Example - API Logging system

def request_logger(func):

    def wrapper():
        print("Request started")
        func()
        print("Request completed")
    return wrapper

@request_logger
def get_customer_data():
    print("Fetching customer data from database")

@request_logger
def place_order():
    print("order placed successfully")

get_customer_data()

place_order()
```
