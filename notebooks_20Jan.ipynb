```python
n=int(input("No of students : "))

students=[]

for i in range(n):
    name = input("Enter name : ")
    roll = input("Roll : ")
    marks = input("Marks : ").split(",")
    marks = [int(m.strip()) for m in marks]
    total = sum(marks)
    avg = total//len(marks)
    grade=""
    if avg>=85:
        grade = 'A'
    elif avg>=70:
        grade = 'B'
    elif avg >=50:
        grade = 'C'
    else:
        grade = 'Fail'

    students.append((name,roll,total,avg,grade))
for s in students:
    print(s)
```

```python
status =""

while status!='exit':
    op=input('Enter operation')
    n1=float(input('enter num1 :'))
    n2=float(input('enter num2 :'))
    match op:
        case '+':
            print(n1+n2)
            
        case '-':
            print(n1-n2)

        case '*':
            print(n1*n2)
            

        case '/':
            if n2==0:
                print("Division by zero error :)")
            else:
                print(n1/n2)

        case _:
            print("Enter correct operation :)")
            

    status=input("exit or not : ")
        
            
```
