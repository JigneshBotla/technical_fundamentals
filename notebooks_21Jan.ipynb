```python
# STUDENT RECORD SYSTEM

students = {}

while True:
    print("\n1 add new student")
    print("2 update marks")
    print("3 show grade")
    print("4 display report")
    print("5 exit")

    n = int(input("Select an option: "))

    match n:
        case 1:
            name = input("Enter name of student: ")
            marks = float(input("Enter marks of student: "))

            if marks >= 70:
                grade = 'A'
            elif marks >= 50:
                grade = 'B'
            else:
                grade = 'Fail'

            students[name] = {
                "marks": marks,
                "grade": grade
            }
            print("Student added successfully!")

        case 2:
            name = input("Which student's marks do you want to update? ")

            if name in students:
                marks = float(input("Enter new marks: "))

                if marks >= 70:
                    grade = 'A'
                elif marks >= 50:
                    grade = 'B'
                else:
                    grade = 'Fail'

                students[name]["marks"] = marks
                students[name]["grade"] = grade
                print("Marks updated successfully!")
            else:
                print("Student not found")

        case 3:
            name = input("Which student's grade do you want to see? ")

            if name in students:
                print(f"Grade: {students[name]['grade']}")
            else:
                print("Student not found")

        case 4:
            print("Generating report...")
            with open("report.txt", "w") as file:
                for name, data in students.items():
                    file.write(f"{name} - Marks: {data['marks']} Grade: {data['grade']}\n")
            print("Report saved to report.txt")

        case 5:
            print("Exiting program...")
            break

        case _:
            print("Invalid option")
```

```python
# SET
web=[101,102,103,104,101,105]
mobile = [103,106,107,102]
offline = [108,109,101,107]

web_set=set(web)

mobile_set=set(mobile)
offline_set=set(offline)

all_customers = web_set.union(mobile_set,offline_set)

print('All customer ID\'s :',all_customers)
#common customers b/w web and mobile
common_online_users=web_set.intersection(mobile_set)
print("\nCustomers using both website and mobile app :",common_online_users)

#find customers that visited website but not offline store

online_only = web_set.difference(offline_set)

print("\nCustomers not visiting offline store :",online_only)

all_customers.add(110)
all_customers.discard(104)

print("\nFinal clean customer database:",all_customers)
print("\nTotal unique customers :",len(all_customers))
```

```python
#E-Commerce Price Analytics Pipeline using List Comprehension

# Raw product prices from database
product_prices = [120, 450, 999, 1500, 80, 2500, 300, 60]
# Step 1: Apply 18% GST tax
taxed_prices = [price * 1.18 for price in product_prices]
print("Prices After Tax:")
print(taxed_prices)
# Step 2: Filter premium products (above 500)
premium_products = [price for price in taxed_prices if price > 500]
print("\nPremium Product Prices:")
print(premium_products)
```

```python
#Step 3: Apply festival discount of 10% on premium products
discounted_prices = [price * 0.90 for price in premium_products]
print("\nDiscounted Premium Prices:")
print(discounted_prices)
# Step 4: Round values for billing system
final_prices = [round(price, 2) for price in discounted_prices]
print ("\rebewel Billing Prices:")
print(final_prices)
# Step 5: Generate report-ready format
billing_report = [
f"Product-{i+1} : Rs.{final_prices[i]}"
for i in range(len(final_prices))
]
print("\nBilling Report Output:")
for record in billing_report:
    print(record)
```

```python
orders = []
customer_ids = set()
REPORT_FILE = "report.txt"

def add_order(order_id,customer_id,product,price,quantity):
    order = {
        "order_id":order_id,
        "customer_id":customer_id,
        "product":product,
        "price":price,
        "quantity":quantity
    }
    orders.append(order) 
    customer_ids.add(customer_id)

def calculate_bill(price,quantity,tax_rate=0.18,discount_rate=0.10):
    subtotal = price*quantity
    tax = subtotal*tax_rate
    discount = subtotal*discount_rate if subtotal>1000 else 0
    total = subtotal +tax-discount
    return subtotal,tax,discount,total
```

```python
def process_orders():
    totals = []
    for order in orders:
        subtotal, tax, discount, total = calculate_bill(
        order["price"], order["quantity"]
        )
        totals.append(total)
        print(order["order_id"], total)
    return totals
def save_report(totals):
    with open(REPORT_FILE, "w") as file:
        for i in range(len(orders)):
            file.write(f" {orders[i]['order_id']} : {totals[i]}\n")
        file.write("\nHighest: " + str(max(totals)))
        file.write("\nLowest: " + str(min(totals)))
add_order(101, "C001", "Laptop", 50000, 1)
add_order(102, "COO2", "Mobile", 15000, 2)
add_order(103, "C001", "Headphones", 2000, 1)
totals = process_orders()
save_report(totals)
```

```python
sales={
}
def add_data(region,salesData):
    sales[region]=salesData


def calc_total_sales(region):
    return sum(sales[region]),sum(sales[region])/len(sales[region])

def add_sales_data(region,newdata):
    sales[region].append(newdata)

def generate_report():
    min1=float('inf')
    max1=float('-inf')
    minr=""
    maxr=""
    avg_final=0
    with open('report.txt','w') as f:
        for k in sales:
            sum1,avg1=calc_total_sales(k)
            min1=min(min1,avg1)
            if avg1 == min1:
                minr=k
        
            max1=max(max1,avg1)
            if avg1 == max1:
                maxr=k
            
            avg_final+=avg1
            f.write(f"{k} total sales : {sum1} average sales : {avg1}")
    f.write("\n--------------PERFORMANCE REPORT FILE--------------")
    
    f.write(f"Region {minr} has minimum average sales of :{min1}")
    f.write(f"Region {maxr} has minimum average sales of :{max1}")
            

    f.close()

add_data("Bangalore",[100,200,300])
add_data("Hyderabad",[100,200,300])

print(calc_total_sales("Bangalore"))
add_sales_data("Bangalore",400)
print(sales)
```
