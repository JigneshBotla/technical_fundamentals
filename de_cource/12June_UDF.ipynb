## Higher order functions

```python
%sql
create or replace temporary view order_items as
SELECT * FROM
VALUES
(1, array('smartphone', 'laptop', 'monitor')), (2, array('tablet', 'headphones', 'smartwatch')), (3, array( 'keyboard', 'mouse'))
AS orders (order_id, items); 
```

```python
%sql
select * from order_items
```

```python
%sql
select order_id,
    transform(items, item -> upper(item)) as items
from order_items
```

```python
%sql
select order_id,
    transform(items, item -> upper(item)) as items,
    filter(items, item -> lower(item) like '%smart%') as smart_items,
    exists(items, item -> lower(item) = 'monitor') as has_monitor
from order_items
```

# There is also an "Aggregate" function

```python
%sql
SELECT * FROM VALUES
(1, array (
named_struct('name', 'smartphone', 'price', 699),
named_struct('name','laptop', 'price', 1199),
named_struct('name', 'monitor', 'price', 399)
)) ,
(2, array ( 
named_struct ('name', 'tablet', 'price', 599), 
named_struct ('name', 'headphones', 'price', 199), 
named_struct ('name', 'smartwatch', 'price', 299)
)) ,
(3, array( 
named_struct ('name','keyboard', 'price',89),
named_struct ('name', 'mouse', 'price', 59)
))
```

```python
%sql
CREATE OR REPLACE TEMP VIEW order_items AS
SELECT * FROM VALUES
(1, array (
named_struct('name', 'smartphone', 'price', 699),
named_struct('name','laptop', 'price', 1199),
named_struct('name', 'monitor', 'price', 399)
)) ,
(2, array ( 
named_struct ('name', 'tablet', 'price', 599), 
named_struct ('name', 'headphones', 'price', 199), 
named_struct ('name', 'smartwatch', 'price', 299)
)) ,
(3, array( 
named_struct ('name','keyboard', 'price',89),
named_struct ('name', 'mouse', 'price', 59)
))
AS orders (order_id, items);
```

```python
%sql
select * from order_items
```

```python
%sql
select order_id,
    transform( transform(items, json_obj -> json_obj.name), item_name -> upper(item_name)) as upperitem ,
    transform( transform(items, json_obj -> json_obj.price), item_price -> 1.1*item_price) as taxedprice

from order_items
```

## Map

```python
%sql
CREATE OR REPLACE TEMP VIEW order_item_prices AS
SELECT * FROM VALUES
(1, map ('smartphone', 699, 'laptop', 1199, 'monitor', 399)), 
(2, map('tablet', 599, 'headphones', 199, 'smartwatch', 299)), 
(3, map('keyboard', 89, 'mouse', 59))
AS orders (order_id, item_prices);
```

```python
%sql
select * from order_item_prices
```

```python
%sql
select 
    transform_keys(item_prices, (x,y) ->upper(x)) as item_prices,
    transform_values(item_prices, (x,y) ->y*1.1) as item_prices,
    map_filter(item_prices, (x,y) ->y>500) as greater500
from order_item_prices
```
