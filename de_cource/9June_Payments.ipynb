```python
%sql
create temporary view date_format as
select payment_id,order_id,payment_timestamp,date_format(payment_timestamp,'yyyy-MM-dd') as payment_date, date_format(payment_timestamp,'HH:mm:ss:m') as payment_time from  gizmobox.bronze.payments
```

```python
%sql
select payment_id,order_id,payment_timestamp,date_format(payment_timestamp,'yyyy-MM-dd') as payment_date, date_format(payment_timestamp,'HH:mm:ss:m'),case payment_staus when 1 then "SUCCESS" when 2 then "PENDING" when 3 then "CANCELLED" when 4 then "FAILED" end as payment_staus,payment_method as payment_time from  gizmobox.bronze.payments
```

```python
%sql
create or replace table gizmobox.silver.payments as
select payment_id,order_id,payment_timestamp,cast(date_format(payment_timestamp,'yyyy-MM-dd') as date)as payment_date, date_format(payment_timestamp,'HH:mm:ss:m') as payment_time,case payment_staus when 1 then "SUCCESS" when 2 then "PENDING" when 3 then "CANCELLED" when 4 then "FAILED" end as payment_staus,payment_method from  gizmobox.bronze.payments
```
