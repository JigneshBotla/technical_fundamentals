```python
df1 = spark.sql('select * from gizmobox.bronze.py_customers')
display(df1)
```

```python
df2 = spark.table('gizmobox.bronze.py_customers')
display(df2)
```

```python
df3 = spark.read.table('gizmobox.bronze.py_customers')
display(df3)
```

## Drop null customer ID's

```python
df4 = df2.dropna(subset=['customer_id'])
display(df4.count())
```

## Drop duplicates

```python
df5 = df4.distinct()
display(df5.count())
```

```python
df5 = df4.dropDuplicates()
display(df5.count())
```

## Remove duplicates based on created_timestamp

```python
df6 = df5.groupBy("customer_id").agg({'created_timestamp': 'max'}).withColumnRenamed("max(created_timestamp)", "max_created_timestamp").select('max_created_timestamp','customer_id')
display(df6)
```

```python
df7 = df6.join(df5, (df6.customer_id == df5.customer_id) & (df6.max_created_timestamp == df5.created_timestamp), "inner").select(df5['*'])
display(df7)
```

```python
df7 = df7.select(
    df7.created_timestamp.cast("timestamp"),
    df7.customer_id,
    df7.customer_name,
    df7.date_of_birth.cast('date'),
    df7.email,
    df7.member_since.cast('date'),
    df7.telephone
)
display(df7)
```
