```python
%sql
create or replace table gizmobox.gold.order_summary_monthly as
select customer_id,date_format(transaction_timestamp,"yyyy-MM") as order_month ,count(distinct(order_id)) as total_orders,sum(quantity) as total_items_bought,sum(price * quantity) as total_amount from gizmobox.silver.orders group by customer_id,order_month
```
